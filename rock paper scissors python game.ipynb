```python
from random import randint 
```

```python
option_list=['roak','paper','scissors']
```

```python
computer=option_list[randint(0,2)]
```

```python
player = input("what's your play? Rock, paper or scissors? ") 

if player == computer:
    print("Tie!")

elif player == 'rock':
    if computer == 'paper':
        print("you lose")
    else:
        print("you win")

elif player == 'paper':
    if computer == 'scissors':
        print("you lose")
    else:
        print("you win")

elif player == 'scissors':
    if computer == 'rock':
        print("you lose")
    else:
        print("you win")

else:
    print("your input is not valid. spell correctly")
```
